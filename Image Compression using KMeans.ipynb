```python
import numpy as np
from skimage.io import imread
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = imread('P.jpg')
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3)
```

```python
img_tf = img.reshape(-1,3)
img_tf = img_tf / 255.0
```

```python
kmeans.fit(img_tf)
```

```python
kmeans.labels_
```

```python
new_img = kmeans.cluster_centers_[kmeans.labels_].reshape(2000,2000,3)
```

```python
plt.imshow(new_img)
```
