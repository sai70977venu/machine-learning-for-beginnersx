```python
cr_data = pd.read_csv('Data/credit-risk-data/cs-training.csv', index_col='Unnamed: 0')
```

```python
cr_data.head()
```

```python
cr_data.agg({'age':'mean','NumberOfDependents':'mean'})
```

```python
cr_data.groupby('age').describe()['MonthlyIncome']
```

```python
cr_data['age_bucket'] = cr_data.age.map(lambda x: x//10)
```

```python
cr_data.groupby('age_bucket').agg({'MonthlyIncome':'median', 'age':'count'})
```

```python
cr_data.info()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
h = cr_data.groupby('age_bucket').agg({'MonthlyIncome':'median'})
```

```python
h.index.tolist()
```

```python
h.values.ravel()
```

```python
df = pd.DataFrame({'age':h.index.tolist(), 'meanincome':h.values.ravel()})
```

```python
df
```

```python
plt.bar(x=df.age, height=df.meanincome)
plt.xlabel('Age in multiple of 10')
```
