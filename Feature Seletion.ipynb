```python
from sklearn.feature_selection import SelectKBest, f_classif, f_regression, SelectFromModel
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
selector = SelectKBest(k=2, score_func=f_classif)
```

```python
selector.fit_transform(iris.data,iris.target)
```

```python
help(SelectKBest)
```

```python
selector = SelectKBest(k=2, score_func=f_regression)
```

```python
selector.fit_transform(iris.data[:,:3], iris.data[:,3])
```

```python
from sklearn.linear_model import LogisticRegression
select = SelectFromModel(LogisticRegression())
```

```python
select.fit_transform(iris.data, iris.target)
```
