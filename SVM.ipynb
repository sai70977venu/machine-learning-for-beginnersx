```python
from sklearn.datasets import make_circles, make_blobs
```

```python
X,Y = make_circles(n_samples=1000, noise=.05)
```

```python
X.shape
```

```python
Y.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,s=10)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
```

```python
svc.fit(X,Y)
```

```python
svc.score(X,Y)
```

```python
svc = SVC(kernel='poly',degree=2)
```

```python
svc.fit(X,Y)
```

```python
svc.score(X,Y)
```

```python
svc.support_vectors_.shape
```

```python
X,Y = make_blobs(n_features=2,centers=2,n_samples=1000,cluster_std =3)
```

```python
plt.scatter(X[:,0],X[:,1],c=Y,s=10)
```

```python
svc = SVC(kernel='poly')
```

```python
svc.fit(X,Y)
```

```python
svc.score(X,Y)
```

```python
svc = SVC(kernel='rbf')
```

```python
svc.fit(X,Y)
```

```python
svc.score(X,Y)
```
