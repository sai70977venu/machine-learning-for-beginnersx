```python
from sklearn.datasets import make_blobs
```

```python
X,Y = make_blobs(n_features=2, n_samples=1000, centers=4)
```

```python
X[1]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.scatter(X[:,0],X[:,1],s=10)
```

```python
from sklearn.cluster import MeanShift
```

```python
ms = MeanShift(n_jobs=-1)
```

```python
ms.fit(X)
```

```python
c = ms.labels_
```

```python
plt.scatter(X[:,0],X[:,1],s=10,c=c)
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(n_jobs=-1, eps=.7)
```

```python
db.fit(X)
```

```python
c = db.labels_
```

```python
plt.scatter(X[:,0],X[:,1],s=10,c=c)
```
